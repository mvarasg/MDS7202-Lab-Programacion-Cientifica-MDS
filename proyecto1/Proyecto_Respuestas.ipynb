## 4. Preprocesamiento

### 4.1 Declarar `ColumnTransformer`

```python
df.info()
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import Pipeline
import pandas as pd


transformations = [('passthrough', 'passthrough', ['id','fraud_bool']),
               ('escalar', StandardScaler(), ['prev_address_months_count','current_address_months_count',
                                              'customer_age','days_since_request',
                                              'intended_balcon_amount','zip_count_4w','velocity_6h',
                                              'velocity_24h','velocity_4w','bank_branch_count_8w',
                                              'credit_risk_score','bank_months_count','has_other_cards',
                                              'proposed_credit_limit','foreign_request','session_length_in_minutes',
                                              'keep_alive_session','device_fraud_count','month',
                                              'x1','x2','name_email_similarity',
                                              'date_of_birth_distinct_emails_4w','email_is_free','device_distinct_emails_8w',
                                              'phone_home_valid','phone_mobile_valid']),
               ('ordinal', OneHotEncoder(), ['payment_type','employment_status','housing_status',
                                             'source','device_os','segmentacion_etaria'])]
```

### 4.2 Transformar datos

```python
# Crear el ColumnTransformer
ct = ColumnTransformer(transformers=transformations)

# Aplicar las transformaciones al DataFrame
df_transformed = ct.fit_transform(df)
```

```python
df_transformed = pd.DataFrame(df_transformed)
df_transformed
```
