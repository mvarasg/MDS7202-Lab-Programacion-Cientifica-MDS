### 1.3.4 Tabla de Resumen [1 punto]

Para generar la tabla de resumen:

- [ ] Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average).
- [ ] Hacer una simplificaci칩n a trav칠s de un mapeo (investigar el m칠todo `map()`) de la variable `parental level of education`  seg칰n la siguiente conversi칩n: 

        some high school -> school
        some college -> school
        high school -> school
        bachelor's degree -> college
        associate's degree -> college
        master's degree -> postgraduate

  Los resultados de este mapeo deben ser guardados en la columna `simple parental level of education`.


- [ ] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener el promedio de las notas. 
- [ ] Agregar seg칰n 2 niveles: `race/ethnicity` y `simple parental level of education` para obtener un conteo de los alumnos en cada grupo y agregarlos como una nueva fila count.
- [ ] Obtener el porcentaje de alumnos con respecto al total. Los porcentajes deben ser strings que contienen la frecuencia de cada grupo con respecto al total y deben ser terminados en '%'. 


Utilizar la tabla de resultados esperados como gu칤a para desarrollar este punto.

````python
def get_resumen(dataframe):
    dict_education = {"some high school": "school",
                      "some college": "school",
                      "high school": "school",
                      "bachelor's degree": "college",
                      "associate's degree": "college",
                      "master's degree": "postgraduate"}

    df = dataframe.copy()
    #Calcular el promedio de las notas y guardarlo en una variable `GPA` (grade point average)
    df['GPA'] = df[df.columns[-3:]].mean(axis=1)
    #Hacer una simplificaci칩n a trav칠s de un mapeo (investigar el m칠todo `map()`) de la variable `parental level of education`
    df["simple parental level of education"] = df["parental level of education"].map(dict_education)

    df_mean = df.groupby(['race/ethnicity', "simple parental level of education"])[['math score', 'reading score', 'writing score','GPA']].mean().reset_index().round(2)
    df_count = df.groupby(['race/ethnicity', "simple parental level of education"]).size().reset_index(name='count')
    df_res = df_mean.copy()
    df_res['count'] = df_count['count']
    len_df = len(df)
    df_res['percentage'] = df_res.apply(lambda x: str(round((x['count']/len_df)*100,2))+' %', axis=1)
    df_res = df_res.set_index(['race/ethnicity', "simple parental level of education"])

    return df, df_res
df_grades , df_resumen = get_resumen(df_grades)
````

> **Justificaci칩n:**


La uni칩n de los dataframes depende del tipo de Join:
- **Outer**: Combina todos los elementos posibles y conserva todo el resto en filas independientes. El resultado es una uni칩n de los dos `dataframes` y logran intersectarse se agregan en otras filas. El resultado tiene m치s de 875 alumnos, por lo tanto no es el resultado que se busca.
- **Left**: Conserva solo los elementos que se hayan combinado correctamente provenientes dataset `left`. El resultado es la uni칩n del primer dataframe (`df_grades`) con la interesecci칩n de este con el `df_other_grades`. Esto asegura que se tendr치n al menos `len(df_grades)` filas. Este resultado es el esperado, dado que se tienen 875 estudiantes y se cruzan tomando el nombre para ver si son iguales.
- **Right**: Conserva solo los elementos que se hayan combinado correctamente provenientes dataset `right`. El resultado es similar al caso anterior, pero con el datafrane derecho, es decir, se obtiene un dataframe con los datos de `df_other_grades` y la intersecci칩n entre este y `df_grades`. La cantidad de filas es `len(df_grades)`, donde muchos alumnos no tienen notas en algunas asignaturas.
- **Inner**: Combina los elementos que se encuentren en ambas tablas. Descarta todo el resto. El resultado es la intersecci칩n entre ambos dataframes, por lo que depender치 de si uno est치 contenido dentro de otro (en este caso, tomando la columna `names`). En este caso se tiene que `df_grades` $\subset$ de `df_other_grades`, por lo que el resultado ser치 el mismo que con Left Join. En efecto, veamos que:

````python
left_join.equals(inner_join)
````

Por lo tanto, en este caso nos debemos quedar con **Inner Join** o **Left Join**. Estrictamente, si se quisiera quedarse con las otras asignaturas y hacer un an치lisis de estas es mejor quedarse con el Inner Join (dado que puede que un df no est칠 contenido completamente en el otro). En cambio, si s칩lo nos interesan las primeras asignaturas y las nuevas son solo un extra (y puede que no tengan notas en estas), es mejor quedarse con el Left Join.
